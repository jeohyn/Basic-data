```
import tensorflow as tf
import pandas as pd
```

using reshape(not flattern)


```
#data
(ind, dep),_=tf.keras.datasets.mnist.load_data()
print(ind.shape, dep.shape)

#make data form to table
ind=ind.reshape(60000, 784) #784=28*28
#one-hot encoding
dep=pd.get_dummies(dep)
print(ind.shape, dep.shape)
```

```
#create model
X=tf.keras.layers.Input(shape=[784])
#hidden layer(84 nodes)-automatic feature extraction(784->84)
H=tf.keras.layers.Dense(84, activation='swish')(X)
Y=tf.keras.layers.Dense(10, activation='softmax')(H)
model=tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```
#fit
#because bach_size=32, 60000/32=1875 so adjust weight 1875times per epoch
model.fit(ind, dep, epochs=10)
```

```
#use model
pred=model.predict(ind[0:5])
#print values prettier
pd.DataFrame(pred).round(2)
```

```
#answer
print(dep[0:5])
```

using flattern(not reshape)

```
#data
(ind, dep),_=tf.keras.datasets.mnist.load_data()
print(ind.shape, dep.shape)

#one-hot encoding
dep=pd.get_dummies(dep)
print(ind.shape, dep.shape)
```

```
#create model
X=tf.keras.layers.Input(shape=[28, 28])
#this code takes the place of reshape()
H=tf.keras.layers.Flatten()(X)
#hidden layer(84 nodes)-automatic feature extraction(784->84)
H=tf.keras.layers.Dense(84, activation='swish')(H)
Y=tf.keras.layers.Dense(10, activation='softmax')(H)
model=tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```
#fit
#because bach_size=32, 60000/32=1875 so adjust weight 1875times per epoch
model.fit(ind, dep, epochs=10)
```

```
#use model
pred=model.predict(ind[6:10])
#print values prettier
pd.DataFrame(pred).round(2)
```

```
#answer
print(dep[6:10])
```
